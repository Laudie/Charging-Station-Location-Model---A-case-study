```python
import gurobipy as gb 
from gurobipy import GRB
```

```python
# Flusso = [5,2,18,30]
# Nodes = [11,21,37,4] 
# Path = [0,1,2,3]
# MATRICE DI INCIDENZA NODI NEI PATH
#        nodi  11  21  37  4   = x11, x21, x37, x4 
#   path    
#   [ 0        1   1   1  1 ] = 5*y0
#   [ 1        0   0   0  1 ] = 2*y1
#   [ 2        1   1   0  0 ] = 18*y2
#   [ 3        0   0   1  0 ] = 30*y3

# max 5y0 + 2y1+ 18y2 + 30y3
# subject to:
#   x11+x21+x37+x4 >= y0
#   x4 >= y1
#   x11+x21 >= y2
#   x37 >= y3
#   x11+x21+x37+x4 = p (NUM STAZIONI DI RICARICA MASSIME)
```

```python
# Flusso è la lista di tutti i flussi di ogni path q
Flusso = [5,2,18,30]
```

```python
# num path q in Q
Path = [0,1,2,3]
```

```python
# tutti i nodi di tutti i paths Q
Nodes = [11,21,37,4] 
```

```python
# dizionario: per ogni path ho la lista dei nodi che passano per quel path
Nodi_Per_Path = {0: [11,4,21,37],
                1: [4],
                2: [11,21],
                3: [37]}
```

```python
# Numero di flussi e nodi
num_flussi = len(Flusso)
num_nodi = max(int(node) for path in Nodi_Per_Path.values() for node in path) + 1

```

```python
# Creazione del modello
model = gb.Model()
```

```python
# Creazione delle variabili di decisione
x = model.addVars(Nodes, vtype=GRB.BINARY, name='x')
y = model.addVars(Path, vtype=GRB.BINARY, name='y')
```

## Objective function

```python
# Definizione della funzione obiettivo
obj_expr = gb.LinExpr()
for i in range(num_flussi):
    obj_expr += Flusso[i] * y[i]
model.setObjective(obj_expr, GRB.MAXIMIZE)
```

```python
model.write('esempio.lp')
```

## Constraints

Num sulle charging station
$$
\sum_{k \in N} x_{k} = p \;\; 
$$

constraint sui nodi 

$$
\sum_{k \in N_q} x_{k} \ge y_q \;\; \forall q \in Q
$$


```python
# Vincoli
for i in range(num_flussi):
    for node in Nodi_Per_Path[i]:
        model.addConstr(x[int(node)] >= y[i])
model.addConstr(x.sum() == 3)
```

```python
model.update()
```

```python
model.write('esempio.lp')
# Risoluzione del modello
model.optimize()
```

```python
if model.status == GRB.OPTIMAL:
    print("NODI IN CUI ATTIVARE LE CHARGING STATION: ")
    for nodo in Nodes:
        x_value = x[nodo].x
        print(f"Nodo {nodo}: x_{nodo}: {x_value}")
else:
    print("Nessuna soluzione ottimale trovata.")
```
